```python
def compute_mean_std_ci(data):
    # Compute the mean and standard deviation along the row axis
    mean_array = np.mean(data, axis=0)
    std_array = np.std(data, axis=0)

    # Compute the shaded area (confidence interval)
    confidence_level = 0.95  # Choose the desired confidence level
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=data.shape[0] - 1)
    margin_error = critical_value * (std_array / np.sqrt(data.shape[0]))
    lower_bound = mean_array - margin_error
    upper_bound = mean_array + margin_error

    return mean_array, std_array, lower_bound, upper_bound
```
